```python
#import path from pathlibrary
#import csv library
from pathlib import Path
import csv
#import pdb; pdb.set_trace()

#set the path for import the file
csvpath = Path('../Resources/budget_data.csv')

#initialzing dictionary to perform the needed analysis
data_analysis = {}

#open the file as object
with open(csvpath, "r") as csvfile: 
    
    #pass csv file to csv function with comma as delimiter
    csvreader = csv.reader(csvfile,delimiter=",")
    
    #read the header
    csv_header = next(csvreader)
    #print(csv_header)
    
    #Initialize variables and paramets
    pnl =  0
    month_count = 0
    pnl_month_list = []
    pnl_list = []
    
    
    #read each line of data
    for row in csvreader:
       #get pnl and month count
        pnl = pnl + int(row[1])
        month_count = month_count+1
        pnl_month_list.append(row[0])
        pnl_list.append(int(row[1]))
        
    #printing the data to check
    print(f'Total months: {month_count}')    
    print(f'Total Profits: {pnl}')
    #print(round(pnl/month_count),2)
    #print(pnl_month_list)
    #print(pnl_list)
    
    #Initailize variables and lists to calculate the daily change
    previous_pnl = 0
    current_pnl = 0
    pnl_diff_list = []
    a = []
    pnl_diff = 0
    
    #Iterating through the list to calculate the differences
    for i in pnl_list:
        current_pnl = i
        if (previous_pnl == 0):
            previous_pnl = current_pnl
        else:
            pnl_diff = current_pnl - previous_pnl #Calculate diff
            previous_pnl = current_pnl  #Assign current value to previous value
            pnl_diff_list.append(pnl_diff)
            a =  pnl_diff_list  
            
    #Intialize variables to calulate average of the differences
    counter = len(a)
    total = sum(a)
    avg = round(total/counter,2) #calculate average
    
    print(f'Average Change: ${avg}')
    data_analysis = dict(zip(pnl_month_list, pnl_diff_list)) #converting list to dictionary
    
    
    #Fetch the greatest profit and lost values from dictionary
    print(f'Greatest Increase in Profits: {max(data_analysis,key = data_analysis.get)} $({max(pnl_diff_list)})')
    print(f'Greatest Decrease in Profits: {min(data_analysis,key = data_analysis.get)} $({min(pnl_diff_list)})')
    #print(min(data_analysis,key = data_analysis.get))
    #https://stackoverflow.com/questions/13704860/zip-lists-in-python 
    #https://www.w3resource.com/python-exercises/dictionary/python-data-type-dictionary-exercise-15.php
   
   
    # Set output file name
output_path = '../Resources/FinancialAnalysis.txt'

    # Open the output path as a file object
with open(output_path, 'w') as file:
    # Write financical analysis to the output file, convert to string
    file.write(f'Financial Analysis\n')
    file.write(f'----------------------------\n')
    file.write(f"Total months: ${month_count}\n")
    file.write(f"Total: ${pnl}\n")
    file.write(f"Average Change: ${avg}\n")
    file.write(f"Greatest Increase in Profits: {max(data_analysis,key = data_analysis.get)} $({max(pnl_diff_list)})\n")
    file.write(f"Greatest Decrease in Profits: {min(data_analysis,key = data_analysis.get)} $({min(pnl_diff_list)})\n")
    
    
#References
#https://stackoverflow.com/questions/13704860/zip-lists-in-python 
#https://www.w3resource.com/python-exercises/dictionary/python-data-type-dictionary-exercise-15.php
#Class Activities https://nu.bootcampcontent.com/NU-Coding-Bootcamp/nu-chi-fin-pt-07-2021-u-c/-/tree/master/01-Lesson-Plans/02-Python
#https://docs.python.org/3/library/pdb.html
```
